# Coverage

[Documentation](https://coverage.readthedocs.io/en/6.1.2/)

In order to gather data about the percent of coverage your tests are providing, you can use a package like `coverage`. `coverage` is a 3rd party package, so it should be installed with `pip`:

```shell
pip install coverage
```

Use coverage run to run your test suite and gather data. However you normally run your test suite, you can run your test runner under coverage.

```shell
# unittest
coverage run -m unittest

# pytest
coverage run -m pytest
```

To limit coverage measurement to code in the current directory, and also find files that weren’t executed at all, add the `--source=.` argument to your coverage command line.


```shell
# unittest
coverage run --source=. -m unittest

# pytest
coverage run --source=. -m pytest
```

Use coverage report to report on the results:

```shell
coverage report
```
